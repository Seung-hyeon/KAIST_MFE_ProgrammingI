## Stock Price Simulation

We often simulate the stock price as a **Geometric Brownian Motion**.

$$
    dS_t = \underbrace{\mu S_t dt}_{\text{drift term}} + \sigma S_t dW_t
$$
where $W_t$ is a Wiener process or Brownian motion.

Many financial assets such as bonds, stocks, options can be simulated based on the Geometric Brownian Motion. To learn more about the theories of mathematical properties, see [here](https://en.wikipedia.org/wiki/Brownian_model_of_financial_markets).

```python
def GBM(mu, sigma, N):
    dt = 0.01
    S = [1]
    for t in range(N):
        dW = dt*np.random.randn()
        S_t = S[-1]*(1 + mu*dt + sigma*dW)
        S.append(S_t)
    return S
```

```python
plt.plot(GBM(0.1, 1, 1000))
plt.plot(GBM(0.1, 2, 1000))
plt.plot(GBM(0.2, 1, 1000))
plt.show()
```

## Interest Rate Simulation

To simulate the interest rate, **Vasicek Model** is the most famous one.

$$
    dr_t = \underbrace{a(b-r_t)}_{\text{mean revert}} dt + \sigma dW_t
$$
where $W_t$ is a Wiener process or Brownian motion.

```python
def Vasicek(a, b, sigma, N):
    dt = 0.01
    r = [1]
    for t in range(N):
        dW = dt*np.random.randn()
        r_t = r[-1] + a*(b-r[-1])*dt + sigma*dW
        r.append(r_t)
    return r
```

```python
plt.plot(Vasicek(1, 1, 1, 2000))
plt.plot(Vasicek(1, 1.5, 1, 2000))
plt.plot(Vasicek(2, 1, 1, 2000))
plt.show()
```

**Asian Option**

* Payoff : Average underlying price over some period $T$
* Feature : Low volatility, Reduce risk of market manimpulation
* Asian call with fixed strike:
$$
    C = \max \left(\mathbb{E}\left[S_t | t\in [0,T]\right] - K, 0 \right)
$$

```python
def AsianCall(T, K, N):
    A = []
    for i in range(N):
        C = max(np.mean(GBM(0.1,2,T*100)) - K, 0)
        A.append(C)
    return np.mean(A)

AsianCall(10, 2, 1000)
```

## Strategy

We often call the **trading strategy** as a method of evaluating portfolio (which is a weighting vector of assets) using a certain information at time $t$. This can be interpreted as a function as,

$$
    f:\mathcal{I_t} \rightarrow \mathbb{R}^n
$$

$f(\mathcal{I_{t-1}}) = \mathbf{w_{t-1}} \rightarrow r_t = \mathbf{r_t}^T\mathbf{w_{t-1}}$

$f(\mathcal{I_{t}}) = \mathbf{w_{t}} \rightarrow r_{t+1} = \mathbf{r_{t+1}}^T\mathbf{w_{t}}$

$f(\mathcal{I_{t+1}}) = \mathbf{w_{t+1}} \rightarrow r_{t+2} = \mathbf{r_{t+2}}^T\mathbf{w_{t+1}}$

where $f$ is a strategy, $\mathbf{w_t}$ is a portfolio vector at time $t$, and $\mathcal{I_t}$ is an information set (or [filtration](https://en.wikipedia.org/wiki/Filtration_(mathematics)) at time $t$.

We can add some constraints to the portfolios.

* $\mathbf{w_t}^T \mathbf{1} = 1$ : Full investment constraints
* $\mathbf{w_t}^T \mathbf{1} = 0$ : Long-short (Market neutral) constraints
* $\|\mathbf{w_t}\|_1 \le c$ : Gross-exposure constraints
* $\|\mathbf{w_t}\|_\infty \le c$ : Maximum Weight (Diversification) constraints

...

** [EX 1] Momentum Strategy **

Buying stocks that have had high returns over the past 3~12 months (<font color='blue'>Winners</font>), and selling those that have had poor returns over the same period (<font color='red'>Losers</font>).

<center>
<img src="https://cssanalytics.files.wordpress.com/2014/06/momentum.jpg" alt="Drawing" style="width: 700px;"/>
</center>

* $\|\mathbf{w_t}\|_1 \le 1$ : Gross-exposure constraints

```python
I = df.pivot(index='date', columns='permno', values='MA (1 year)')
ret = df.pivot(index='date', columns='permno', values='ret')
holding = 1
```

```python
def f(t):
    ranking = I.iloc[t-holding].rank(pct=True)
    w = (ranking > 0.9)*1 - (ranking < 0.1)
    w /= np.abs(w).sum()
    return w
```

```python
r = pd.Series()
for t in range(14, 70):
    r[ret.iloc[t].name] = ret.iloc[t].fillna(0).dot(f(t))
```

```python
r.cumsum().plot()
plt.title('momentum 1~12 month')
plt.show()
```

** [EX 2] Momentum Strategy (Long only) **

Buying stocks that have had high returns over the past 3~12 months (<font color='blue'>Winners</font>)

<center>
<img src="https://cssanalytics.files.wordpress.com/2014/06/momentum.jpg" alt="Drawing" style="width: 700px;"/>
</center>

* $\mathbf{w_t}^T \mathbf{1} = 1$ : Full investment constraints

```python
def f(t):
    ranking = I.iloc[t-holding].rank(pct=True)
    w = (ranking > 0.9)
    w /= w.sum()
    return w

def f_M(t):
    w = np.ones(I.iloc[t-holding].shape[0])
    w /= w.sum()
    return w
```

```python
r = pd.Series()
r_M = pd.Series()
for t in range(14, 70):
    r[ret.iloc[t].name] = ret.iloc[t].fillna(0).dot(f(t))
    r_M[ret.iloc[t].name] = ret.iloc[t].fillna(0).dot(f_M(t))
```

```python
(1+r).cumprod().plot(label='momentum')
(1+r_M).cumprod().plot(label='market')
plt.title('momentum (long) 1~12 month')
plt.legend()
plt.show()
```

Consider Transaction Cost (Buy, sell symmetric)

$$
    \text{TC}_t = \gamma V_t \|\mathbf{w_{t-1}} - \mathbf{w_t}  \|_1
$$
where $V_t$ is an amount of wealth (or portfolio value) at time $t$.

In Korea, $\gamma \approx 0.8\%$

```python
gamma = 0.008
```

```python
r = pd.Series()
r_M = pd.Series()
w_ = 0
for t in range(14, 70):
    w = f(t)
    r_TC = gamma*np.linalg.norm(w-w_, ord=1)
    r[ret.iloc[t].name] = ret.iloc[t].fillna(0).dot(w) - r_TC
    r_M[ret.iloc[t].name] = ret.iloc[t].fillna(0).dot(f_M(t))
```

```python
(1+r).cumprod().plot(label='momentum')
(1+r_M).cumprod().plot(label='market')
plt.title('momentum (long) 1~12 month (w/ TC)')
plt.legend()
plt.show()
```
